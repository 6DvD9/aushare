```python
#clean price collections: remove invalid files
import aushare.stock.fundamental as td
import os
import pandas as pd

#file_delist = 'data/ASXDelistcodes.csv'
removeList =[]
codelist = td.getAllASXListCode().values
for code in codelist:
    file_name = 'data/'+str(code)+'_price.csv'
    if os.path.isfile(file_name):
        if os.stat(file_name).st_size <10*1024:
            os.remove(file_name)
            removeList.append(code)
    else:
        removeList.append(code)
d = {'ASX code': removeList}
df = pd.DataFrame(data=d)
#df.to_csv(file_delist)
```

```python
# clean revenue data collections: remove invalid files

import aushare.stock.fundamental as td
import os
import pandas as pd

file_delist = 'data/ASXDelistcodes.csv'
removeList =[]
codelist = td.getAllASXListCode().values
for code in codelist:
    file_name = 'data/'+str(code)+'_revenue.csv'
    if os.path.isfile(file_name):
        try:
            df = pd.read_csv(file_name,index_col =0)
            if (df.empty):
                os.remove(file_name)
                removeList.append(code)
            else:
                print(df['Total Revenue'])
        except:
            os.remove(file_name)
            removeList.append(code)
    else:
        removeList.append(code)
d = {'ASX code': removeList}
df = pd.DataFrame(data=d)
with open(file_delist, 'a') as f:
    df.to_csv(f, header=False)
```

```python
#clean price collections: remove invalid files for ETFs
from aushare.etf import cons as ct
from aushare.etf import etfmarketdata as td

import os
import pandas as pd

etffilelist = []
codelist = td.getAllETFListCode().values
for code in codelist:
    file_name = ct.DAILY_PRICE_FILE%code
    if os.path.isfile(file_name):
        if os.stat(file_name).st_size >1*1024:
            etffilelist.append(code)
            
            
print(etffilelist)

with open('etflistfile.txt', 'w') as filehandle:  
      for listitem in etffilelist:
          filehandle.write('%s\n' % listitem)


```
